```python
import requests
from bs4 import BeautifulSoup

# The URL of the page you want to scrape
url = 'https://www.mayoclinic.org/healthy-lifestyle/recipes/diabetes-meal-plan-recipes/rcs-20077150'
base_url = 'https://www.mayoclinic.org'
# Send a GET request to the webpage
response = requests.get(url)
mayo_clinic_recipes = []
# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the response using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all recipe titles based on their HTML structure
    # This is a hypothetical example; you'll need to adjust the selector based on the actual page structure
    recipe_lists = soup.find_all('ol', class_='content')  # Adjust this selector based on the actual HTML
    # # Print each recipe title
    for li in recipe_lists:
        for item in li.find_all('li'):
            if item:
                recipe_item = {}
                recipe_item['recipe_name'] = item.text.strip()
                recipe_item['source_url'] = ''.join([base_url, item.find('a')['href']])
                mayo_clinic_recipes.append(recipe_item)
else:
    print(f"Failed to retrieve content, status code: {response.status_code}")
```

```python
import re

def remove_space_between_number_and_unit(text):
    # Define a regular expression pattern to match the space between number and unit
    pattern = r'(\d+)\s+([a-zA-Z]+)'
    # Replace the space with no space
    replaced_text = re.sub(pattern, r'\1\2', text)
    return replaced_text
```

```python
import time
```

```python
import urllib
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
def scrap_recipe(recipes:list)->list:
    parsed_recipes = []
    error_recipes = []
    
    for recipe in recipes:
        try:
            source_url = recipe['source_url']
            req = urllib.request.Request(source_url)
            req.add_header('User-Agent', 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:106.0) Gecko/20100101 Firefox/106.0')
            req.add_header('Accept', 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8')
            req.add_header('Accept-Language', 'en-US,en;q=0.5')
            response = urllib.request.urlopen(req)
            if response.getcode() == 200:
                response = response.read().decode('utf-8')
                parsed_recipe = dict()
                parsed_recipe['recipe_name'] = recipe['recipe_name']
                parsed_recipe['source_url'] = source_url

                soup = BeautifulSoup(response, 'html.parser')
                recipe_div = soup.find('div', class_='recipes')
                # print(recipe_div)
                ingredient_div = soup.find('ol', class_ = 'bullet')
                ingredients = []
                for item in ingredient_div.find_all('li'):
                    ingredients.append(item.text.strip())
                parsed_recipe['ingredients'] = ','.join(ingredients)
                directions = []
                recipe_h2 = recipe_div.find_all('h2')
                for item in recipe_h2:
                    if item and item.text == 'Directions':
                        next_tag = item.find_next_sibling()
                        while next_tag and next_tag.name == 'p':
                            directions.append(next_tag.text.strip())
                            next_tag = next_tag.find_next_sibling()
                parsed_recipe['directions'] = ','.join(directions)     
            
                nutrients_div = recipe_div.find('div', class_ = 'r-nutrients')
                nutrients_h3_div = nutrients_div.find('h3')
                parsed_recipe['serving_size'] = nutrients_h3_div.text

                nutrients = []
                for item_ul in nutrients_div.find_all('ul'):
                    for item_li in item_ul.find_all('li'):
                        nutrients.append(remove_space_between_number_and_unit(item_li.text.strip()))
                parsed_recipe['nutrients'] = ','.join(nutrients)
                parsed_recipes.append(parsed_recipe)
                time.sleep(1)
            else:
                error_recipes.append(recipe)
        except Exception as e:
            print(e)
            error_recipes.append(recipe)
            time.sleep(1)
            continue 
       
    return parsed_recipes
```

```python
results = scrap_recipe(recipes_source)
```
